#Creating Database Table

```python
import pymysql

db = pymysql.connect(host="127.0.0.1", user="root", password="")
cursor = db.cursor()

# Creating TESTDB
cursor.execute("CREATE DATABASE IF NOT EXISTS TESTDB")
db.select_db("TESTDB")

cursor.execute("DROP TABLE IF EXISTS EMPLOYEE")

sql = """CREATE TABLE EMPLOYEE (
   FIRST_NAME  CHAR(20) NOT NULL,
   LAST_NAME  CHAR(20),
   AGE INT,
   SEX CHAR(1),
   INCOME FLOAT )"""

cursor.execute(sql)
print("The table was created successfully.!")

db.close()
```

#INSERT Operation

```python
import pymysql

db = pymysql.connect(host="127.0.0.1", user="root", password="", database="TESTDB")
# prepare a cursor object using cursor() method
cursor = db.cursor()

# Prepare SQL query to INSERT a record into the database.
sql = """INSERT INTO EMPLOYEE(FIRST_NAME,
   LAST_NAME, AGE, SEX, INCOME)
   VALUES ('Mac', 'Mohan', 20, 'M', 2000)"""

try:
   # Execute the SQL command
   cursor.execute(sql)
   db.commit()  # Commit your changes in the database
   print("Data entered successfully!")
except Exception as e:
   db.rollback() # Rollback in case there is any error
   print("Error:", e)

db.close()
```

#READ Operation

```python
import pymysql

db = pymysql.connect(host="127.0.0.1", user="root", password="", database="TESTDB")
cursor = db.cursor()

# Selecting those with income above 1000
sql = "SELECT * FROM EMPLOYEE WHERE INCOME > 1000"

try:
   cursor.execute(sql)
   results = cursor.fetchall()

   for row in results:
      fname = row[0]
      lname = row[1]
      age = row[2]
      sex = row[3]
      income = row[4]
   # Printing the obtained data
      print ("fname = %s, lname = %s, age = %d, sex = %s, income = %d" % \
             (fname, lname, age, sex, income ))
except:
   print ("Error: unable to fetch data")

db.close()
```

# Update Operation


```python
import pymysql

db = pymysql.connect(host="127.0.0.1", user="root", password="", database="TESTDB")
cursor = db.cursor()

# SQL command to increase age
sql = "UPDATE EMPLOYEE SET AGE = AGE + 1 WHERE SEX = 'M'"

try:
   cursor.execute(sql)
   db.commit()
   print("Data update successful.!")
except:
   db.rollback()
   print("Update failed..")

db.close()
```

# DELETE Operation


```python
import pymysql
# Open database connection
db = pymysql.connect(host="127.0.0.1", user="root", password="", database="TESTDB")
cursor = db.cursor()
# Prepare SQL query to DELETE required records
sql = "DELETE FROM EMPLOYEE WHERE AGE > 20"

try:
   # Execute the SQL command
   cursor.execute(sql)
   # Commit your changes in the database
   db.commit()
   print("Data deletion successful.!")
except:
   db.rollback()
   print("Delete failed..")
# disconnect from server
db.close()
```

#READ Operation


```python
import pymysql

# Open database connection
db = pymysql.connect(host="127.0.0.1", user="root", password="", database="TESTDB")
cursor = db.cursor()

# SELECT command to display all data
sql = "SELECT * FROM EMPLOYEE"

try:
   cursor.execute(sql)
   results = cursor.fetchall()

   # Check if any records were returned
   if len(results) == 0:
      print("No records found in the EMPLOYEE table.")
   else:
      print(f"--- Total Records Found: {len(results)} ---")

   # Print each line individually
   for row in results:
      print(f"Name: {row[0]} {row[1]}, Age: {row[2]}, Sex: {row[3]}, Income: {row[4]}")

except Exception as e:
   print("Error: Unable to fetch data:", e)
# Disconnect from server
db.close()
```
